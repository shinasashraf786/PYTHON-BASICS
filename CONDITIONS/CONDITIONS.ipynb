```python
num = int(input('enter the number'))

if num>0:
  print(num,'positive')
elif num == 0:
  print(num,'zero')
else:
  print(num,'negative')
```

```python
age = int(input('enter the age : '))

if age<10:       #false
  print('lp')

elif age<13:    #true
  print('up')

elif age<15:
  print('high school')

elif age<17:
  print('hs')

else:
  print('invalid')
```

```python
age = int(input('enter the number'))
health = input('enter your health condition 1 for critical,2 for normal,')

if age>60 and health=='2':
  print('you are eligible for second dose')

elif age>60 and health=='1':
  print('you are not eligible for second dose due to bad health condition')

elif age>18 and health=='2':
  print('you are eligible for first dose')

elif age>18 and health=='1':
  print('you are not eligible for first dose')

else:
  print('you are not eligible')

```

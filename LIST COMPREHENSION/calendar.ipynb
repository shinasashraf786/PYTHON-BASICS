```python
# What is Module in Python?
# The module is a simple Python file that contains collections of functions and global variables and with
# having a .py extension file. It is an executable file and to organize all the modules we have the concept called Package in Python.

# Examples of modules:
# Datetime
# Regex
# Random
#calendar
```

```python
# What is Package in Python?
# The package is a simple directory having collections of modules. This directory contains Python modules and.
# The package is simply a namespace. The package also contains sub-packages inside it.

# Examples of Packages:
# Numpy
# Pandas
```

```python
# What is Library in Python
# The library is having a collection of related functionality of codes that allows you to perform many tasks without writing your code.
# It is a reusable chunk of code that we can use by importing it into our program, we can just use it by importing that library and
# calling the method of that library with a period(.). However, it is often assumed that while a package is a collection of modules,
# a library is a collection of packages.

# Examples of Libraries:
# Matplotlib
# Pytorch
# Pygame
# Seaborn
```

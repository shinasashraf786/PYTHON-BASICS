```python
n = int(input('enter how many numbers you want to enter to input'))

odd = []
even = []
dict1 = {}
list1 = []

for i in range(n):
  num = int(input('enter the numbers : '))

  list1.append(num)

  if num%2 == 0:
    even.append(num)
  else:
    odd.append(num)

  dict1[num] = list1.count(num)


print(odd)
print(even)
print(dict1)
print(list1)

#[3,4,5,61,3,46]


```

```python
n = int(input('enter how many numbers you want to enter to input'))

odd = []
even = []
dict1 = {}
list1 = []

for i in range(n):
  num = int(input('enter the numbers : '))

  list1.append(num)

  if num%2 == 0:
    even.append(num)
  else:
    odd.append(num)

  if num not in dict1:
   dict1[num] = 1
  else:
   dict1[num] = dict1[num]+1


print(odd)
print(even)
print(dict1)
print(list1)


#[3,4,5,61,3,46]


```

If the number num is already a key in the dictionary, its value (count) is incremented by 1.

If the number num is not a key in the dictionary, it is added as a key with the value (count) set to 1.

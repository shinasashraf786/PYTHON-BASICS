**Seaborn**

Seaborn is a Python data visualization library built on top of Matplotlib.

It provides a high-level interface for creating aesthetically pleasing statistical graphics.

Seaborn is particularly useful for exploring and visualizing datasets, allowing users to gain insights into their data quickly and effectively.

One of the key features of Seaborn is its ability to create visually appealing plots with minimal code.

It offers a wide range of plot types, including scatter plots, line plots, bar plots, histograms, box plots, violin plots, and more.

```python
import seaborn as sns
```

Box Plot

A box plot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset based on five summary statistics:* minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.*

It is useful for detecting outliers and comparing distributions between different groups or categories

```python
sns.boxplot(df['tip'])
```

```python
sns.boxplot(x='tip',y='day',data=df)
```

```python
sns.boxplot(x='tip',y='day',data=df,hue='smoker')
```

Swarm Plot:

A swarm plot is similar to a strip plot, but it spreads out the points to avoid overlap, giving a better visualization of the distribution of the data. Seaborn's swarmplot() function is used to create swarm plots.

swarmplot is like the scatterplot where points donot overlap

```python
sns.boxplot(x='day',y='total_bill',data=df)
sns.swarmplot(x='day',y='total_bill',data=df,color=".10")
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
sns.swarmplot(x='day',y='total_bill',data=df,color=".10",hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
sns.swarmplot(x='day',y='total_bill',data=df,hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
sns.swarmplot(x='day',y='total_bill',data=df,hue='sex',dodge=True)
```

Violin Plot:

A violin plot combines the features of a box plot and a kernel density estimation plot to visualize the distribution of a numerical variable across different categories.

Seaborn's violinplot() function is useful for comparing the distribution of numerical variables across categories.

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

Pair Plot:
A pair plot creates a grid of scatter plots for all pairs of variables in a

dataset, along with histograms for each variable's distribution. Seaborn's

pairplot() function can be used to create pair plots.

```python
df = sns.load_dataset('flights')

df.head()
```

```python
flights = df.pivot(index="month", columns='year', values='passengers')
sns.heatmap(flights)
```

```python
flights = df.pivot(index="month", columns='year', values='passengers')
sns.heatmap(flights,annot=True)
```

```python
flights = df.pivot(index="month", columns='year', values='passengers')
sns.heatmap(flights,annot=True,fmt='d')
```

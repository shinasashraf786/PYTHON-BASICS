**NUMPY**

class name of numpy is ndarray

Numpy is a Python library used for working with arrays

In Python we have lists that serves the purpose of arrays,but they are

slow to process.

Numpy aims to provide an array object that is up to 50* faster than

traditional Python lists.

The array object in Numpy is called ndarray,it provides a lot of

supporting functions that make working with ndarray very easy

Arrays are very frequently used in data science,where speed and

resources are very important

Collection of homogenous data

It does not support hetrogenous data

size is fixed

ARRAY JOINING


```python
import numpy as np

a = np.array([(3,4,2),(4,5,3)])

b = np.array([(4,3,1),(5,32,0)])

c = np.vstack((a,b))

c
```

```python
import numpy as np

a = np.array([(3,4,2),(4,5,3)])

b = np.array([(4,3,1),(5,32,0)])

c = np.row_stack((a,b))

c
```

```python
import numpy as np

a = np.array([(3,4,2),(4,5,3)])

b = np.array([(4,3,1),(5,32,0)])

c = np.hstack((a,b))

c
```

```python
import numpy as np

a = np.array([(3,4,2),(4,5,3)])

b = np.array([(4,3,1),(5,32,0)])

c = np.column_stack((a,b))

c
```

```python
import numpy as np

a = np.array([(3,4,2),(4,5,3)])

b = np.array([(4,3,1),(5,32,0)])

x = np.concatenate((a,b))

x
```

```python
import numpy as np

a = np.array([(3,4,2),(4,5,3)])

b = np.array([(4,3,1),(5,32,0)])

x = np.concatenate((a,b),axis=1)

x
```

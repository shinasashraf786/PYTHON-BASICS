```python
#function without argument
def function_name():

  a = int(input('enter the number1'))
  b = int(input('enter the number2'))

  return a+b

print(function_name())
```

```python
def sub(a,b):

  if a>b:

    sub = a-b
  else:
    sub = b-a
  return sub

print(sub(1,9))
```

```python
#defualt arguments and non defualt arguments

def sub(a,b=40):

  if a>b:
    sub = a-b

  else:
    sub = b-a
  return sub

print(sub(9))

#non default argument cannot be passed after default argument
```

```python
#defualt arguments and non defualt arguments

def add(a,c,d,b=80):

  add = a+b+c+d
  return add

print(add(9,50,50,23))
```

```python
def factorial(num):

  factorial = 1

  if num<0:

    print('Factorial does not exist for negative numbers')

  elif num==0:

    print('Factorial of 0 is 1')

  else:

    for i in range(1,num+1):


      factorial = factorial * i

  return factorial

print(factorial(5))
```

```python
def validate_pin(pin):

  res = ''
  count = 0
  flag_digi = 0

  for i in pin:
    count = count+1

    if i not in ['1','3','5','7','9']:
      flag_digi = 1

  if flag_digi == 0:
    if count == 4 or count ==6:
      res = 'valid'
    else:
        res = 'invalid'


  else:
    return 'invalid'

  return res

print(validate_pin('1379'))
```

```python
def func(string):

  vowels = 'aeiouAEIOU'

  for i in string:

    if i in vowels:
      string = string.replace(i, '')

  return string

print(func('shinas'))
```

```python


def sub(n1,n2):
  if n1<n2:
    (n1,n2) = (n2,n1)


  return n1-n2


def div(n1,n2):

  if n1 < n2:
    (n1,n2) = (n2,n1)

  return n1/n2

print(sub(4,8))

print(div(10,5))
```

```python
def swap_num(fn):   #decorator function

  def wrapper(n1,n2):

    if n1<n2:

      (n1,n2) = (n2,n1)

    return fn(n1,n2)

  return wrapper

@swap_num
def sub(n1,n2):

  return n1-n2

@swap_num
def div(n1,n2):

  return n1/n2

print(sub(4,12))

print(div(10,5))
```

Identity Operator

is, is not

```python
a = 70
b = 47
print(a is b)
```

```python
a = 70
b = 47
print(type(a) is type(b))
```

```python
a = 70
b = 47
print(type(a) is not type(b))
```

**Membership Operator**

in,not in

```python
x = 'hello'

print('e' in x)
```

```python
x = 'hello'

print('e' not in x)
```

```python
x = 'hello how are you'

print('how' in x)
```

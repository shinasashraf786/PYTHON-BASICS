The union() method returns a new set with all items from both sets.

```python
names = {'shinas','aswanth','pranav','abishek'}
names1 = {'shinas','aswanth','pranav','abishek','ajay'}

names3 = set.union(names,names1)
print(names3)
```

The difference() method will return a new set that will contain only the items from the first set that are not present in the other set.

```python
names = {'ajay','shinas','aswanth','pranav','abishek','rahul'}
names1 = {'shinas','aswanth','pranav','abishek','gokul'}

names3 = set.difference(names1,names)
print(names3)
```

```python
names = {'shinas','aswanth','pranav','abishek'}
names1 = {'shinas','aswanth','pranav','abishek','ajay'}

names3 = set.intersection(names1,names)
print(names3)
```

```python
names = {'shinas','aswanth','pranav','abishek',}
names1 = {'shinas','aswanth','pranav','abishek','ajay'}

names3 = set.isdisjoint(names1,names)
print(names3)
```
